#### Consider a chess knight moving on the first quadrant of the plane. It starts at (0,0), and at each step will move two units in one direction and one unit in the other, such that x≥0 and y≥0. At each step the knight randomly selects a valid move, with uniform probability. For example, from (0,1), the knight will move to (1,3), (2,2), or (2,0), each with probability one-third.



1. After 10 moves, what is the expected Euclidean distance of the knight from the origin?

```python
# compute the possible coordinates and euclidean distances with uniform probability
def func():
    k = 100
    array1 = [[0, 0]]
    coor = []
    knight = [[1, 2], [2, 1], [-1, 2], [-2, 1], [1, -2], [2, -1], [-1, -2], [-2, -1]]
    
    for j in range(1,k):
        print(j)
        for i in range(len(array1)):
            for l in range(len(knight)):
                x = array1[i][0]+knight[l][0]
                y = array1[i][1]+knight[l][1]
                if x >= 0 and y >= 0:
                    array = [x, y]
                    if array not in coor: 
                        coor.append(array)
        array1 = coor[:]
        coor = []
    
    prob = {}
    dist = {}
    dist_2 = {}
    for i in range(len(array1)):
        for l in range(len(knight)):
            x = array1[i][0]+knight[l][0]
            y = array1[i][1]+knight[l][1]
            if x >= 0 and y >= 0:
                array = [x, y]
                array_string = str(array)
                if array not in coor:
                    coor.append(array)
                    dist[array_string] = (x**2 + y**2)
                if array_string in prob:
                    prob[array_string] += 1
                else:
                    prob[array_string] = 1

    total = sum(prob.values())
    prob = {k: prob[k] / total for k in prob}
    print(total) 
    print()
    print(prob)
    print()
    print(dist)
    print()
    
    expected = 0
    expected_2 = 0
    for k in prob:
        expected += prob[k] * dist[k]**(1/2)
        expected_2 += prob[k] * dist[k]
    sd = (expected_2 - expected**2)**(1/2)
    
    print("E1", expected)
    print("E2", expected_2)
    print("SD", sd)
    
    
func()
```
